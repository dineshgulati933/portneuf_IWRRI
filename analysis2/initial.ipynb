```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
aoi = gpd.read_file('shp/portneuf3_clean/portneuf3_clean.shp')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
aoi.plot(ax=ax, legend=True, cmap='Set3', edgecolor='black', linewidth=0.5)

ax.axis('off')

plt.tight_layout()
# plt.savefig('pneuf_3p.png', dpi=200)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
aoi.plot(ax=ax, legend=True, cmap='Set3', edgecolor='black', linewidth=0.5)

# Label each polygon with UID at its centroid
for idx, row in aoi.iterrows():
    centroid = row.geometry.centroid
    ax.text(centroid.x, centroid.y, str(row['UID']), fontsize=14, ha='center', va='center')

ax.axis('off')
plt.tight_layout()
plt.savefig('pneuf_3p.png', dpi=200)
plt.show()
```
