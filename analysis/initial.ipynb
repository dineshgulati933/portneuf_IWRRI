```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
huc12 = gpd.read_file('portneuf_huc12/portneuf_huc12.shp')
```

```python
huc12.info()
```

```python
huc12.plot( figsize=(10, 10), legend=True, cmap='Set3')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
huc12.plot(ax=ax, legend=True, cmap='Set3', edgecolor='black', linewidth=0.5)

ax.axis('off')

plt.tight_layout()
plt.savefig('huc12.png', dpi=200)
plt.show()
```
